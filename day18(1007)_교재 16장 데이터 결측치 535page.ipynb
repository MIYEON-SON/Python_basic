## 워드클라우드 시각화

```python
# 549-550page
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
file_name='C:/Users/ASIA-16/Desktop/Digital_dev/data/book_example_data/littleprince_djvu.txt'

with open(file_name) as f:
    text=f.read()
```

```python
print(text)
```

```python
# wordcloud 의 이미지를 생성

wc_img=WordCloud().generate(text)

# plt를 통한 wordcloud 표현
plt.imshow(wc_img, interpolation='bilinear')
plt.show()
```

```python
# wc_img2
wc_img=WordCloud(background_color='white',max_font_size=300,\
                width=800,height=400).generate(text)
plt.imshow(wc_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```
