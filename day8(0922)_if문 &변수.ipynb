# if문의 필요성

> "돈이 있으면 택시를 타고, 돈이 없으면 걸어간다."

```python
money=True

if money:
    print('택시 타세요.')
else:
    print('걸으세요.')
```

### 비교연산자

```python
x=2
y=3
print(x<y)
print(y<x)
print(x==y)
print(x!=y)
```

```python
# 택시비를 상정- 1
money=10000000
if money >=3800:
    print("택시")
else:
    print("걷기")
```

```python
# and or not 을 활용한 if문
money=0
pay=True
if money >= 3800 or pay:
    print("택시")
else:
    print("걷기")
```

```python
# in과 not in 살펴보기

1 in [1,2,3] # 리스트, 튜플, 문자열이 in 이후에 올 수 있음
```

```python
1 in [10,20,30]
```

```python
100 not in [10,20,30]
```

```python
# 주머니가 있고 그 속에 있는 물건으로 택시비를 지불한다고 상정

pocket=['wallet','phnoe','card','airpod','watch']
money=0

if money>=3800 or 'card' in pocket:
    print('택시')
else:
    print('걷자')
```

### 2. copy 모듈을 이용

* def + def =class
* class + class = module
* module + module = package
* package + package = library

```python
#    패키지       모듈
from copy import copy
a=[1,2,3]
b=copy(a)
```

## 변수를 만드는 여러 가지 방법 

```python
# 1번째 방법
a, b = ('python', 'life')
print(a)
print(b)
```

```python
# 2번째 방법
(a, b) = 'python', 'life'
print(a)
print(b)
```

```python
# 3번째
[a,b]=['python','life']
print(a)
print(b)
```

```python
# 여러 변수에 같은 값을 대입
a=b=c='python'
```

```python
# 간단한 변수 바꾸기
a='A'
b='B'
a,b=b,a
a
```
