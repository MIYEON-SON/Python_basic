```python
# iloc 및 loc의 비교

test_data={'Name':['dongjin','minsoo','daeho','boyoung'],
           'Year':[2016,2017,2012,2006],
           'Gender':['Male','Male','Male','Female']
          }
test_df=pd.DataFrame(test_data, index=['a','b','c','d'])
test_df
```

```python
# iloc - positional indexing (위치 기반 인덱싱)

print(test_df.iloc[0,0])
#print(test_df.iloc[0,'Name'])  -> 의도적 오류1
#print(test_df.iloc['a',0]  -> 의도적 오류2
```

```python
# loc - Label based indexing (명칭 기반 인덱싱)

print(test_df)
print(test_df.loc['a','Name'])
print(test_df.loc[:,'Name'])
```

```python
# 명칭 기반 loc와 위치 기반 iloc 의 결과 확인
print('명칭 기반 loc slicing\n',test_df.loc['a':'b','Name'])
print('위치 기반 iloc slicing\n',test_df.iloc[0:2,0])
```
