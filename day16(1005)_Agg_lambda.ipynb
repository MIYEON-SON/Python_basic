## 정렬, Aggregation 함수, Groupby 적용

#### Dataframe, Series의 정렬 - sort_values()

DataFrame과 Series의 정렬을 위해서는 sort_values( ) 메서드를 이용한다. sort_values()는 RDBMS SQL의 order by키워드와 유사하다. sort_values()의 주요 입력 파라미터는 by, ascending, inplace이다. by로 특정 칼럼을 입력하면 해당 칼럼으로 정렬을 수행한다. ascending=True로 설정하면 오름차순으로 정렬하며, ascending=False로 설정하면 내림차순으로 설정된다. 기본(default)는 ascendling=True이다. inplace도 default는 inplace=False이다.

```python
titan_df.sort_values(by=['Name'])[:2]
```

여러 개의 칼럼으로 정렬하려면 by에 리스트 형식으로 정렬하려는 칼럼을 입력하면 된다. 다음의 예는 Pclass와 Name으로 내림차순으로 정렬한 결과이다. 

```python
titan_df.sort_values(by=['Pclass','Name'],ascending=False)
```

### groupby() 적용

DataFrame의 groupby()는 RDBMS SQL의 groupby 키워드와 유사하면서도 다른면이 있기에 주의해야 한다.DataFrame의 groupby사용시 입력 파라미터 by에 칼럼을 입력하면 대상 칼럼으로 groupby된다. DataFrame에 groupby를 호출하면 DataFrameGroupBy라는 또다른 형태의 DataFrame을 반환한다.

#### groupby(by='Pclass')를 호출하면 Pclass 기준으로 GroupBy된 DataFrame Groupby객체를 반환한다. 

```python
# Pclass를 기준으로 Groupby하고, 전체 데이터를 count()한 후 Cabin과 Gender 컬럼만 호출하자
titan_df.groupby(by='Pclass').count()[['Cabin','Gender']]
```

```python
# Pclass를 기준으로 Cabin과 Gender 컬럼을 호출하자
titan_df.groupby(by='Pclass')[['Cabin','Gender']].count()  # 위와 동일한 결과
```

```python
# Pclass를 기준으로 Groupby한 후 수치형 변수인 Age, Fare의 최소/최대값을 알아보자
titan_df.groupby(by='Pclass')[['Age','Fare']].agg([min,max])
```

```python
## 서로 다른 agg 함수를 적용
## Age에는 max, SibSp에는 sum, Fare에는 mean
agg_def={'Age':[max,min],'SibSp':sum,'Fare':'mean'}
titan_df.groupby('Pclass').agg(agg_def)
```

### 결측치 처리

판다스는 결손 데이터(Missing Data)를 처리하는 편리한 API를 제공한다. 결손 데이터는 칼럼에 값이 없는, 즉 NULL인 경우를 의미하며, 이를 넘파이의 NaN으로 표시한다. 기본적으로 머신러닝 알고리즘은 이 NaN값을 처리하지 않으므로 이 값을 다른 값으로 대체해야 한다. 또한 NaN값은 평균, 총합 등의 함수 연산시 제외가 된다. 특정 칼럼의 100개 데이터 중 10개가 NaN 값일 경우, 이 칼럼의 평균 값은 90개 데이터에 대한 평균이다.  

```python
# 결측치의 여부 확인 - isna()
titan_df.isna().sum()
```

### fillna()로 결손 데이터 대체하기

```python
# 가장 많은 빈도값:: B96 B98
titan_df.Cabin.value_counts()
```

```python
# 그러므로 우리는 'C000'으로 해당 데이터들을 imputation(대치)한다
titan_df['Cabin']=titan_df['Cabin'].fillna('C000')

# Age와 Embarked 데이터도 대치
titan_df['Age']=titan_df['Age'].fillna(titan_df.Age.mean())
```

```python
## Q1 해당 데이터의 소수자리를 둘째 자리로 바꿔 보자
## np.round(data,2)를 활용
```

```python
titan_df['Embarked']=titan_df['Embarked'].fillna('S')

# 결측치 없음 확인
titan_df.isna().sum()
```

### apply lambda 식으로 가공

판다스는 apply함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공한다. 판다스의 경우 칼럼에 일괄적으로 데이터 가공을 하는 것이 속도 면에서 더 빠르나 복잡한 데이터 가공이 필요할 경우엔 어쩔 수 없이 apply lambda를 이용한다. 먼저 lambda 식에 익숙하지 않을 분들을 위해 설명을 아래와 같이 하겠다.

```python
def get_square(x):
    result=x**2
    return result
```

```python
print('9의 제곱은:', get_square(9))
```

```python
# list에 lambda 적용
a=[1,2,3]
crux=map(lambda x: x**2, a)
list(crux)
```

```python
# pandas df에 lambda 적용
titan_df['Name_len']=titan_df['Name'].apply(lambda x: len(x))
```

```python
# 50세 이상 old, 50세 이하 young

titan_df['Young_Old']=titan_df['Age'].apply(lambda x: 'young' if x<=50 else 'old')
titan_df.Young_Old.value_counts()
```

```python
# 해당 Young_Old 컬럼에서 우리가 Young 수를 세어 보자.
titan_df[titan_df['Young_Old']=='young'].count()
```

```python
# 25세 이상 adult, 나머지 Adult, 60세 이상 elderly

titan_df['Young_Old']=titan_df['Age'].apply(lambda x: 'young' if x<=25 else ('adult' if x<=60 else 'elderly'))
titan_df.Young_Old.value_counts()
```

```python
# Category에 따른 분류
def age_cat(df_age):
    char=''
    if df_age<=5: char='baby'
    elif df_age<=13: char='kid'
    elif df_age<=19: char='teen'
    elif df_age<=25: char='student'
    elif df_age<=40: char='yougn_adult'
    elif df_age<=60: char='adult'
    else: char='elderly'
    
    return char
```

```python
titan_df['Age_cat']=titan_df['Age'].apply(lambda x: age_cat(x))
```
