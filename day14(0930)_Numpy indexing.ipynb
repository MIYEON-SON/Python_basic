### 넘파이 인덱싱

```python
import numpy as np
array1d=np.arange(1,10)
array2d=array1d.reshape(-1,3)
print(array2d)
```

```python
# 1. 특정 데이터 추출
print('row=2, col=0','\n',array2d[2,0])
print('row=0, col=2','\n',array2d[0,2])

# 2. slicing indexing
print('row=:1, col=2:3','\n',array2d[0:1,2:3])  # 둘 다 :(슬라이싱)이면 2차원

# 3. fancy indexing
print('array2d[[0,1],1]=>','\n',array2d[[0,1],1])
print('array2d[[0,1],1:2]=>','\n',array2d[[0,1],1:2])
```
